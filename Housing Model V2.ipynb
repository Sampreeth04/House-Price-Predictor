```python
#DATA PROCESSING
```

```python
#Converting date strings into year and month
for i in range(len(df['date'])):
    df['date'][i] = int(df['date'][i][0:6])   
```

```python
#Cutting the date into quarters, data will be sorted from 1-8
bins = [201401, 201403, 201406, 201409, 201412, 201503, 201506, 201509, 201512]
labels = ['Quarter 1', 'Quarter 2', 'Quarter 3', 'Quarter 4','Quarter 5', 'Quarter 6', 'Quarter 7', 'Quarter 8']

df['date'] = pd.cut(df['date'], bins=bins, labels=labels)
```

```python
#Mapping the data into numerical values
df['date'] = df['date'].map({'Quarter 1': 1, 'Quarter 2': 2, 'Quarter 3': 3, 'Quarter 4': 4, 'Quarter 5': 5, 'Quarter 6': 6, 'Quarter 7': 7,'Quarter 8': 8})
```

```python
#Cutting year build and renovated, data will be sorted with 20 year separations from 1900 to 2020
bins = [-1]
labels = [0]
a = 1
for i in range(1900, 2040, 20):
    bins.append(i)
for i in range(1900, 2020, 20):
    labels.append(a)
    a += 1
df['yr_built'] = pd.cut(df['yr_built'], bins=bins, labels=labels)
df['yr_renovated'] = pd.cut(df['yr_renovated'], bins=bins, labels = labels)
```

```python
#Dropping unused columns and outliers
df.drop(['id'], axis=1, inplace=True)
df.drop(15870, inplace=True)
df.drop(12777, inplace=True)
```

```python
#Data is seen to be measured on different ranges, so we are scaling the data to increase model accuracy
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()

df_scaled = pd.DataFrame(data=standard_scaler.fit_transform(df), columns=df.columns)
```

```python
#MODEL BUILDING
```

```python
#Splitting train test data
from sklearn.model_selection import train_test_split
X = df_scaled.drop(['price'], axis=1)
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10)
```

```python
#Linear Regression model
from sklearn.linear_model  import LinearRegression
lin_reg_model = LinearRegression()

lin_reg_model.fit(X_train, y_train)
predictions = lin_reg_model.predict(X_test)
```

```python
#Evaluation Report
#69% accuracy is okay, but there is room for improvement
from sklearn.metrics import mean_absolute_error, mean_squared_error
print('Absolute mean error - ', mean_absolute_error(y_test, predictions))
print('Mean square error - ', mean_squared_error(y_test, predictions))
print('Root mean square error - ', np.sqrt(mean_squared_error(y_test, predictions)))
print('Percentage accuracy: {}%'.format(round(int(lin_reg_model.score(X_test, y_test)*100))))

lin_reg_accuracy = round(int(lin_reg_model.score(X_test, y_test)*100))
```

```python
#Gradient Booster Model
#92% accuracy is awesome!
from sklearn import ensemble
gradient_booster_model = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2, learning_rate = 0.1)

gradient_booster_model.fit(X_train, y_train)
gradient_booster_predictions = gradient_booster_model.predict(X_test)

print('Absolute mean error - ', mean_absolute_error(y_test, gradient_booster_predictions))
print('Mean square error - ', mean_squared_error(y_test, gradient_booster_predictions))
print('Root mean square error - ', np.sqrt(mean_squared_error(y_test, gradient_booster_predictions)))
print('Percentage accuracy: {}%'.format(round(int(gradient_booster_model.score(X_test, y_test)*100))))

gradient_booster_accuracy = round(int(gradient_booster_model.score(X_test, y_test)*100))
```

```python
#CONCLUSION
```

```python
model_df = pd.DataFrame({'Model Name' : ['Linear Regression', 'Gradient Booster'],
                         'Accuracy' : [lin_reg_accuracy, gradient_booster_accuracy]})
model_df.sort_values(by='Accuracy', ascending=False)
```

```python
#In conclusion, Gradient Booster is the most accurate model for the given dataset with an accuracy of 92%
```
